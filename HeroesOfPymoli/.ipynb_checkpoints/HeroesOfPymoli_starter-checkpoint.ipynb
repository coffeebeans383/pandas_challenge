### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_data.head()
```

```python
player_data = len(purchase_data["SN"].value_counts())
players_total = pd.DataFrame({"Total Players":[player_data]})
players_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculations for unique items
unique_items = len((purchase_data["Item ID"]).unique())
# average price of each items
average_price = (purchase_data["Price"]).mean()
#find number of purchases
total_purchases = (purchase_data["Purchase ID"]).count()
#calulate total revenue
total_revenue = (purchase_data["Price"]).sum()

# Create data frame with obtained values
purchase_summary_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                           "Average Price":[average_price], 
                           "Number of Purchases": [total_purchases], 
                           "Total Revenue": [total_revenue]})

purchase_summary_df["Average Price"] = purchase_summary_df["Average Price"].map("${0:,.2f}".format)
purchase_summary_df["Total Revenue"] = purchase_summary_df["Total Revenue"].map("${0:,.2f}".format)
purchase_summary_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
genderDemo_data = purchase_data.groupby(["Gender"]).nunique(["SN"]).sort_values(["SN"],ascending=False).
percentage_gender = genderDemo_data["SN"] / 576 * 100
percentage_gender
genderDemo_df = genderDemo_data.drop(columns=["Purchase ID","Age","Gender","Item ID","Item Name","Price"])
genderDemo_df.rename(columns={"SN":"Total Count"},inplace=True)
genderDemo_df["Percentage of Players"]= percentage_gender
genderDemo_df["Percentage of Players"] = genderDemo_df["Percentage of Players"].map("{0:,.2f}%".format)
genderDemo_df.index.name = None
genderDemo_df.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PurchaseAnalysis_data = purchase_data.groupby(["Gender"])
purchase_count = PurchaseAnalysis_data["Purchase ID"].count()
purchase_count
# Average purchase prices by gender
avg_price = PurchaseAnalysis_data["Price"].mean()

# Average purchase  by gender 
total_rev = PurchaseAnalysis_data["Price"].sum()

# Average purchase total by gender divivded by purchase count by unique shoppers
avg_purchase = total_rev/genderDemo_df["Total Count"]

# Create data frame with obtained values 
PurchaseAnalysis_df = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": avg_price,
                                    "Average Purchase Value":total_rev,
                                    "Avg Purchase Total per Person": avg_purchase})

PurchaseAnalysis_df.style.format({"Average Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})
```

## Age Demographics

```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=group_names)
purchase_age_data=purchase_data.groupby("Age Group")

#Calculate the numbers and percentages by age group
age_count = purchase_age_data["SN"].nunique()

# Calculate percentages by age category 
age_perc = (age_count/576) * 100

#Create a summary data frame to hold the results
age_dem_df = pd.DataFrame({"Total Count": age_count,"Percentage of Players": age_perc})
age_dem_df.index.name = None
# Optional: round the percentage column to two decimal points
age_dem_df=age_dem_df.style.format({"Percentage of Players":"{:,.2f}%"})

# Display Age Demographics Table
age_dem_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#count purchases by age group
purchase_count = purchase_age_data["Purchase ID"].count()
#purchase_count
# average purchase price by age group 
avg_price_age = purchase_age_data["Price"].mean()

#total purchase by age group 
total_purchase_age = purchase_age_data["Price"].sum()

SN_Unique = purchase_age_data["SN"].nunique().count()

# average purchase per person in age group 
avg_purchase_SN = total_purchase_age/SN_Unique

# Create data frame with obtained values
age_purchase_df = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price": avg_price_age,
                                 "Total Purchase Value":total_purchase_age,
                                 "Average Purchase Total per Person": avg_purchase_age)

# Format the data frame with no index name in the corner
age_purchase_df.index.name = None

# Format with currency style
age_purchase_df.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by screen names
groupBy_SN = purchase_data.groupby("SN")

#total purchases by SN
purchase_count_SN = groupBy_SN["Purchase ID"].count()

#verage purchase by SN 
avg_price_SN = groupBy_SN["Price"].mean()

# total purchases by SN 
total_rev_SN = groupBy_SN["Price"].sum()


# Create data frame with obtained values
top_spenders = pd.DataFrame({"Purchase Count": purchase_count_SN,
                             "Average Purchase Price": avg_price_SN,
                             "Total Purchase Value":total_rev_SN})

# Sort in descending order to obtain top 5 spender names 
top_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head(5)

# Format with currency style
top_spenders=top_spenders.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create new data frame with items related information 
popularItems_data = purchase_data[["Item ID", "Item Name", "Price"]]
popularItems_data

# Group the item data by item id and item name 
popularItems_data = popularItems_data.groupby(["Item ID","Item Name"])

# Count the number of times an item has been purchased 
item_count = popularItems_data["Price"].count()

# Calcualte the purchase value per item 
item_rev = (popularItems_data["Price"].sum()) 

# Find individual item price
item_price = item_rev/item_count

# Create data frame with obtained values
popularItems_df = pd.DataFrame({"Purchase Count": item_count, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":item_rev})
popularItems_df = popularItems_df.sort_values(["Purchase Count"], ascending=False).head(10)

# Format with currency style
popularItems_df.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitableItems_df = popularItems_df.sort_values(["Total Purchase Value"], ascending=False).head(10)

profitableItems_df.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
